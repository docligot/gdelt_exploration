```python
from newspaper import Article
import nltk
```

```python
def extract_newsdata(x):
    try:
        news = Article(x)
        news.download()
        news.parse()
        news.nlp()
        output = pd.DataFrame([[news.url, news.authors, news.text, news.summary, news.keywords]], \
                              columns=['url', 'authors', 'text', 'summary', 'keywords'])
        return output
    except:
        return
    
def extract_from_urls(x):
    output = pd.DataFrame(columns = ['url', 'authors', 'text', 'summary', 'keywords'])
    for i in x['sourceurl']:
        output = output.append(extract_newsdata(i), ignore_index=True)
    return output
```

```python
extract_newsdata(urls.sourceurl[1482]).to_csv('extract.csv')
```
